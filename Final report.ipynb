### Business Problem

I decided to take a hypothetical situation and thought about the company that wanted to expand its business and was looking for some business advice.

The background was the following:

US based company Bicycle Inc is looking for the new expansion opportunities. Since the company has a strong presence and great coverage of its shops in the United States, it wants to tackle new markets. Therefore, the company has decided to open its first bike shop in Europe. The company wants to know where to open its first shop. 

In order to deliver insightful recommendations, certain citeria had to be considered. Therefore, some important questions needed to be answered: 
Does the city has large enough population?
What is the climate in that city? Is it possible to ride a bike all year round?
What is the percentage of the population using the bike? Is cycling popular in that city?
What about the competitors - is there a large penetration of bike shops?

This information is considered to be useful not only for the companies selling bikes but also for other companies promoting an active lifestyle - selling sporting clothes, sporting equipment, supplements, etc. In addition, enterprises renting sporting equipment might also be interested in such information. 


### Data

In order to solve the business problem, the following data was used:

Foursquare API - bike shops. This data was projected to provide useful insights on the best place in the city to open the first bike shop. Number of bike shops was taken as a rough guidance on overall competition in the city. And even though large number of bike shops could create more competition, it also meant that this area was more likely to receive higher number of clients.

Population of EU capitals (capitals of the country are the largest cities in the country, therefore, only population in capitals was taken).

Bicycle usage. The statistics on bicycle usage was taken from the European Cyclists’ Federation page (link: https://ecf.com/resources/cycling-facts-and-figures). Even though some of the data was rather outdated, it still worked as a good guidance on the general usage of bicycles in the city.

Climate conditions. Average temperature in the cities was taken from Wikipedia page (link: https://en.wikipedia.org/wiki/List_of_cities_by_average_temperature#Europe) and yearly sunshine hours was taken from https://en.wikipedia.org/wiki/List_of_cities_by_sunshine_duration#Europe
